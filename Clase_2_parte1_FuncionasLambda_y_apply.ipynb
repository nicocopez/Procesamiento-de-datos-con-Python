## Funciones y funciones lambda


### Escribir funciones con def

Esta es la manera estandard de definir funciones que tenemos en Python. Nos permite asignarles un nombre y utilizarlas en distintas partes del código.

Ejemplo: Escribir la función elevar cuadrado.

```python
# Ejemplo funcion elevar al cuadrado:

def elevar_al_cuadrado(x): # Definimos la función de un argumento x
    resultado = x**2 # Calculamos el argumento x al cuadrado
    return resultado # Con return imponemos que la función de un argumento nos devuelva el resultado
```

```python
print('Ejemplo usando def: ' + str(elevar_al_cuadrado(2)))
```

### Funciones lambda

Las expresiones lambda, también llamadas funciones anónimas, se usan cuando queremos hacer algo simple y y rápido en una linea en lugar de nombrar formalmente una función.

Son una forma corta de declarar funciones pequeñas y anónimas (no es necesario proporcionar un nombre para las funciones lambda) pero que se comportan como funciones normales declaradas con la palabra clave def. Haciendo esto, consigues escribir menos lineas de código, lo que puede ser especialmente útil cuando tienes que mantener programas grandes.

Es una manera adicional de definir funciones que nos será muy útil en el manejo de DataFrames.


Ejemplo: Reescribir la función elevar al cuadrado como una función lambda:

```python
elevar_al_cuadrado_con_lambda = lambda x: x ** 2

print( 'Ejemplo usando lambda: ' + str(elevar_al_cuadrado_con_lambda(2)))

## Nota: Aquí hemos asignado un nombre a la función lambda, aunque recordamos que lo habitual será
## utilizarlas en contextos donde no sea necesario utilizar nombres
```

**Interludio 1:** Repaso bucles for:

Los bucles son una herramienta para alterar el flujo normal de un programa. Nos permiten repetir una porción de código tantas veces como indiquemos. Python incluye únicamente dos tipos de bucle: while y for. 

Ejemplo: Devolver valores en una lista:

```python
colores = ["rojo", "verde", "negro", "azul"]

for color in colores:
    print(color)
```

Ejemplo: devolver números en un rango:

```python
for i in range(1, 4):
    print(i)
```

Si además de obtener el valor queremos obtener su posición que ocupa en la lista (es decir, el índice), tenemos esta opción:

```python
for idx, val in enumerate(colores):
    print(idx, val)
```

**Interludio 2:** Comprensión de listas:

```python
colores = ["rojo", "verde", "negro", "azul"]

# ejemplo 1
[color for color in colores]
```

## Método apply

Una alternativa a usar loops para iterar sobre un DataFrame is usar la función apply() de pandas.
El método apply nos permite transformar columnas de un DataFrame o crear columnas nuevas de una forma organizada y compacta.
Su funcionamiento consiste en aplicar una determinada función sobre la columna a crear o modificar. Esa función debe ser un argumento de apply.

La función a aplicar podrá ser definida por una función lambda que es posible que sólo se utilice en esa línea concreta de código.


Ejemplo: Sumar dos a todos los elementos en un dataframe, haciendo uso de def y de una función lambda:

```python
# Ejemplo de data frame:
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}) # diccio
print('Data frame: ')
df
```

```python
# Funcion para sumar dos:
def add_2(x):
    return x+2
```

```python
# Sumar dos a todos los elementos del df con apply
df_a = df.apply(add_2)
print('Data frame tras hacer apply con la función sumarle dos: ')
df_a
```

```python
# Como hemos visto, también se puede hacer usando una función lambda:
df_b = df.apply(lambda x: x + 2)
print('Data frame tras hacer apply con una función lambda: ')
df_b
```

Y si queremos mantener las columnas originales? Podemos crear nuevas columnas:

```python
# Si lo que queremos es que aparezcan como columnas nuevas:
df[['D', 'E', 'F']] = df.apply(lambda x: x + 2)
df
```

Que pasa si ahora queremos aplicar la función a una sola columna?

```python
# Volvemos a definir nuestro data frame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
df
```

```python
# Ahora aplicamos la función sumar dos solo a la columna A:
df['A'] = df['A'].apply(add_2)
print('Data frame tras sumar dos a la columna A: ')
df
```

```python
# Como hemos visto, también se puede hacer usando una función lambda:
df_b = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
df_b['A'] = df_b['A'].apply(lambda x: x + 2)
print('Data frame tras hacer apply con una función lambda: ')
df_b
```

Y si queremos que guarde la información en una nueva columna?

```python
# Volvemos a definir nuestro data frame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
print('Data frame: ')
df
```

```python
# Para crear una nueva columna, tenemos dos opciones:
#df['D'] = df['A'].apply(lambda x: x + 2) # opción 1 
df['D'] = df.apply(lambda x: x['A'] + 2, axis = 1) # opción 2

print('Data frame tras hacer apply con una función lambda: ')
df
```
