## Repaso: Listas y diccionarios

**Listas** <br>
Las listas son como su propio nombre indica una lista de datos en los que hay orden. Que haya orden, significa que se tiene en cuenta la posición en la que esta cada elemento. Combiene recordar que el primer elemento de una lista en Python es el 0 y no el 1. Se crean escribiendo sus elementos entre corchetes

```python
# Ejemplos de lista:
mi_lista = [5, 6, 7, 8, 9]
print('Ejemplo de lista numérica:' + str(mi_lista))

mi_lista_b = ["verde", "rojo", "azul", "rosa", "blanco", "amarillo"]
print(mi_lista_b)
```

```python
#Ejemplo, como seleccionar un elemento:
mi_lista_b[3]
```

Algunas operaciones habituales con listas son:

- mi_lista[i]: Devuelve el elemento que está en la posición i de la lista. <br>
- mi_lista.pop(i): Devuelve el elemento en la posición i de una lista y luego lo borra. <br>
- mi_lista.append(elemento): Añade elemento al final de la lista. <br>
- mi_lista.insert(i, elemento): Inserta elemento en la posición i. <br>
- mi_lista.extend(lista2): Fusiona lista con lista2. <br>
- mi_lista.remove(elemento): Elimina la primera vez que aparece elemento. <br>

**Diccionarios**<br>
En Python, un diccionario consta de palabras que tienen asociado algo, como en los diccionarios convencionales. Se construyen escribiendo sus elementos entre llaves {"palabra 1" : "definicion 1", "palabra 2" : "definicion 2",...}. A las "palabras" se les denominan keys y values a sus "definiciones".  No puede haber dos keys iguales, pero sí dos values iguales. Al contrario que las listas, los diccionarios no tienen orden.



```python
# Ejemplo de diccionario:
mi_diccionario = {'Color 1':'verde','Color 2':'rojo','Color 3':'azul','Color 4':'naranja'}
print(mi_diccionario)
```

Igual que en los diccionarios convencionales a cada palabra se puede dar varios significados, en los diccionarios de Python, por cada key puede haber varios elementos. Esto se puede conseguir escribiendo una lista por cada elemento.

```python
# Ejemplo:
mi_diccionario = {'Colores':['verde','rojo','azul','naranja'], 
                  'Formas': ['cuadrado', 'circulo', 'triangulo', 'rectangulo'],
                 'Tamano': ['grande', 'mediano', 'pequeno']}
print(mi_diccionario)

```

Algunas operaciones habituales con diccionarios son:

- mi_diccionario.get('key'): Devuelve el valor que corresponde con la key introducida. <br>
- mi_diccionario.pop('key'): Devuelve el valor que corresponde con la key introducida, y luego borra la key y el valor.<br>
- mi_diccionario.update({'key':'valor'}): Inserta una determinada key o actualiza su valor si ya existiera. <br>
- 'key' in mi_diccionario: Devuelve verdadero (True) o falso (False) si la key (no los valores) existe en el diccionario.<br>
- 'definicion' in mi_diccionario.values(): Devuelve verdadero (True) o falso (False) si definición existe en el diccionario (no como key).<br>

## Pandas


Pandas es una librería de código abierto de Python para la manipulación y el análisis de datos. Ofrece estructuras de datos y operaciones para manipular tablas numéricas y series temporales. Su nombre viene de "PANel DAta"

Pandas Dataframe: Conjunto de datos tabulados en dos dimensiones en el cual se puede guardar datos de distintos tipos (caractéres, enteros, valores de punto flotante,etc.). Es similar a una hoja de cálculo o una tabla de SQL.

Pandas Series: Es un array 1D capaz de contener cualquier tipo de datos. Una Serie representa una columna, que puede ser independiente o pertenecer a un Pandas Dataframe.

Para trabajar con la librería pandas, primero tenemos que importarla:

```python
# Importamos la librería pandas con el nombre corto pd:
import pandas as pd
```

Hay muchas formas diferentes de crear un dataframe.<br>

Por ejemplo vamos a crear un dataframe a partir de un diccionario:

```python
# Creamos un diccionario hecho de listas:
mi_diccionario = {'A':['1', '2', '3'],'B':[4,5,6],'C':[7,8,9]}

# Creamos el dataframe a partir del diccionario
df = pd.DataFrame(mi_diccionario)

# Veamos que pinta tiene:
df
```

Tambien podemos crear un dataframe a partir de una lista de listas:

```python
# Creamos un diccionario a partir de una lista de listas
df = pd.DataFrame([[1,2,3], [4,5,6], [7,8,9]], columns=['A','B','C'], index = ['row 1','row 2','row 3'])
# Nota, fíjate como se pueden nombrar las columnas y las filas.
df
```

### Tipos de variable:

#### Variables numericas:
**Variables Discretas**: <br>
Son también llamadas discontinuas, y están asociadas a conteos o enumeraciones, razón por la cual, solo permiten ser registradas con números enteros (0,1,2,3, etc.)<br>
**Variables Continuas**: <br>
Son aquellas características que son medidas dentro de un rango continuo infinito de valores numéricos y se registran con números reales. Pueden presentar cualquier valor dentro de cierto intervalo.<br>


#### Variables categoricas:
**Variable Nominal**: <br>
Es aquella variable cualitativa cuya categoría no sigue ningún orden, se agrupa sin ninguna jerarquía entre sí.<br>
**Variable Ordinal**:<br>
Son aquellas variables categóricas con orden secuencial o progresión natural esperable o jerarquía.

### Funciones de acceso

Los DataFrames son estructuras indexadas por filas y por columnas. Se pueden utilizar estos índices para acceder a los contenidos del dataset que almacenan usando el mecanismo de slicing, exactamente igual que en las listas y los arrays de Numpy (mi_array[fila,columna]. Veamos algunos ejemplos:

```python
df # Este era nuestro dataframe
```

```python
# Obtenemos la primera fila:
df[0:1]
```

```python
# Obtenemos la segunda fila:
df[1:2]
```

```python
# Obtenemos las dos primeras filas:
df[0:2]
```

```python
# Obtenemos la primera columna:
df['A']
```

```python
# Obtenemos las dos primeras columna:
df[['A', 'B']]
```

```python
# Obtenemos el tercer elemento de la segunda columna:
df['B'][2:3] # El orden es indiferente, df[2:3]['B'] tambien valdría
```

El resultado de este acceso es una Serie, no un valor. En la mayoría de los casos nos va a interesar acceder al valor puro, con su correspondiente tipo, y no obtenerlo almacenado en un DataFrame de una sola fila y columna. Para ello, usaremos las siguientes funciones de acceso a valores:

- **df.loc** Acceso por etiqueta, siendo la etiqueta el índice (fila o columna). 
- **df.iloc** Acceso por posición entera. 
- df.at Acceso rápido por etiqueta, orientado a un solo elemento. 
- df.iat Acceso rápido por posición entera, orientado a un solo elemento. 
- df.ix Acceso por etiqueta o por posición. Se considera retirado (deprecated) porque resulta confuso.

Ejemplo: Obtener el tercer elemento de la segunda columna con .loc

```python
df.loc['row 3','B']
```

#### Diferencias entre .loc, .iloc, .at y .iat

- iloc y loc pueden seleccionar multiples files y/o columnas. Potencialmente para hacer vectorizaciones.
- iat y at solo pueden acceder a valores escalares. Es decir, no valen para hacer slicings. Son la versión rápida de iloc y loc.


- at y loc solo trabajan con indices.
- iat y iloc solo trabajan con posiciones.



Las funciones de acceso nos valen para seleccionar elementos y trabajar con ellos, pero tambien para reemplazarlos.
